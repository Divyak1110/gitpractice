Q12. Create a class isosceles_right_triangle which inherits from isosceles_triangle and
right_triangle.


```python
class Triangle:
    number_of_sides = 3
    def __init__(self, angle1,angle2,angle3):
        self.angle1 = angle1
        self.angle2 = angle2
        self.angle3 = angle3
    def check_angles(self):
        if (self.angle1 + self.angle2 + self.angle3) == 180:
            print('Angles belongs to the traingle')
            return True
        else:
            print('Angles does not belong to the traingle')
            return False
class equilateral_triangle(Triangle):
    def e_traingle(self):
        if self.angle1 == self.angle2 == self.angle3 == 60:
            return 'It is an equilateral_triangle'
        return 'It is not an equilateral_triangle'
        
class right_triangle(Triangle):
    def r_triangle(self):
        if self.angle1 == 90 or self.angle2 == 90 or self.angle3 == 90:
            return 'It is right_traingle'
        return 'It is not right_traingle'
class isosceles_triangle(Triangle):
    def i_traingle(self):
        if self.angle1 == self.angle2 or self.angle2 == self.angle3 or self.angle1 == self.angle3:
            return 'It is an isosceles_triangle'
        return 'It is an isosceles_triangle'
class isosceles_right_triangle(right_triangle,isosceles_triangle):
    def ir_traingle(self):
        if self.r_triangle() == 'It is right_traingle' and \
           self.i_traingle() == 'It is an isosceles_triangle':
            return 'it is an isosceles and right_triangle.'
        return 'it is not an isosceles and right_triangle.'

```
