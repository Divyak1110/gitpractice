Q1. Using Python script as a calculator
1. Employee class has 3 data members EmployeeID, Gender (String), Salary and PerformanceRating(Out of 5) of type int. It has a get() function to get these details from the user.
2. JoiningDetail class has a data member DateOfJoining of type Date and a function getDoJ to get the Date of joining of employees.
3. Information Class uses the marks from Employee class and the DateOfJoining date from the JoiningDetail class to calculate the top 3 Employees based on their Ratings and then Display, using readData, all the details on these employees in Ascending order of their Date Of Joining.








```python
from datetime import datetime
class Employee:
    def __init__(self):
        self.EmployeeId = 0
        self.Gender = " "
        self.Salary = 0
        self.PerformanceRating = 0
    def get_details(self):
        self.EmployeeId = int(input("Enter you Employeeid:  "))
        self.Gender = input("Enter Gender(Male/Female): ")
        self.Salary = int(input("Enter your Salary: "))
        self.PerformanceRating = int(input("Enter Rating (1-5): "))
class JoiningDetail():
    def __init__(self):
        self.DOJ = None
    def date_of_joining(self):
        date_str = input("Enter DOJ (YYYY-MM-DD): ")
        self.DOJ = datetime.strptime(date_str, "%Y-%m-%d").date()
class Information(Employee,JoiningDetail):
    def __init__(self):
        Employee.__init__(self)
        JoiningDetail.__init__(self)
    def read_data(self):
        print("Id is:", self.EmployeeId,
             "Gender is:",self.Gender,
             "Salary is:",self.Salary,
             "Performance rating is:", self.PerformanceRating,
             "Date of Joining is:", self.DOJ)

n = int(input("How many employees do you want to enter? "))
list = []
for i in range(n):
    print("\nEnter Employee details:", (i+1))
    emp = Information()
    emp.get_details()
    emp.date_of_joining()
    list.append(emp)
top_3 = sorted(list, key=lambda x: x.PerformanceRating)[:3]
top_3_sorted_by_date = sorted(top_3, key=lambda x: x.DOJ)
print("\n--- Top 3 Employees (By Rating, Sorted by Date) ---")
for emp in top_3_sorted_by_date:
    emp.read_data()
```
