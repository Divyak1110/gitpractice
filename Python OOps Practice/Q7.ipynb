Q7. Create a program in python to demonstrate Polymorphism.
1. Make use of private and protected members using python name mangling techniques.


```python
class Account:
    def __init__(self):
        self.__sbalance = 1000
        self.__cbalance = 0
    def show_balance(self):
        pass

class savings_account(Account):
    def show_balance(self):
        return self._Account__sbalance
class credit_card(Account):
    def show_balance(self):
        super().show_balance()

        return self._Account__cbalance
my_account = [savings_account(),credit_card()]
for acc in my_account:
    print(acc.show_balance())
```
