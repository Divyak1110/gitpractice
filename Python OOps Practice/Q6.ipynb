Q6. Show a basic implementation of abstraction in python using the abstract classes.
1. Create an abstract class in python.
2. Implement abstraction with the other classes and base class as abstract class.


```python
from abc import ABC, abstractmethod
class Bank_Account(ABC):
    def __init__(self, name, Amount):
        self.name = name
        self.Amount = Amount
        self.Balance = 1000
    @abstractmethod
    def update_balance(self):
       pass
class Savings_Account(Bank_Account):
    def update_balance(self):
        if self.Amount > 0:
            self.Balance += self.Amount
        else:
            self.Balance -= abs(self.Amount)
        return ('Account:', self.name ,
                'Type of Account: Savings ',
                'Account Balance:',self.Balance)
class Current_Account(Bank_Account):
    def update_balance(self):
         if self.Amount > 0:
            self.Balance += self.Amount
         else:
            self.Balance -= abs(self.Amount)
         return ('Account:', self.name ,
                 'Type of Account: Current',
                 'Account Balance:',self.Balance)

S_Account = Savings_Account('Mary', 5000)
C_Account = Current_Account('My LLC' ,-1000)
S_Account.update_balance()
C_Account.update_balance()
            
```
