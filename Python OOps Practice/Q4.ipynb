```python
import numpy as np

score = 0
deck_selection = np.random.randint(1, 14, size=8) 

# Start the loop to play through the 8 cards
for i in range(len(deck_selection) - 1):
    current_card = deck_selection[i]
    next_card = deck_selection[i+1]
    
    print(f"\n--- Round {i+1} ---")
    print(f"Current card showing: {current_card}")
    
    user_choice = input("Will the next card be 'higher' or 'lower'? ").lower().strip()
    
    print(f"The next card is: {next_card}")

    if next_card > current_card:
        actual_outcome = "higher"
    else:
        actual_outcome = "lower"

    if user_choice == actual_outcome:
        print("Correct! +20 points.")
        score += 20
    else:
        print("Incorrect. -15 points.")
        score -= 15

print(f"\nGame Over! Your final score is: {score}")

    
        
        
    
    
```
